```python
netflix['type'].value_counts().plot(kind='bar')
plt.grid(axis='y',linestyle = '--',alpha = 0.6)
```

```python
netflix['rating'].value_counts().head(10).plot(kind='bar',colors=['orange'])
plt.grid(axis='y',linestyle = '--',alpha = 0.8)
plt.title("Title Count")
```

```python
netflix['country'].value_counts().head(10).plot(kind='barh')
plt.grid(axis='y',linestyle = '--',alpha = 0.6)
plt.title("Top 10 Countries with Most Netflix Content")
```

```python
netflix[netflix['director'] != 'Unknown']['director'].value_counts().head(10).plot(kind='bar')
plt.title("Top 10 Directors")
```

```python
netflix['type'].value_counts().plot(kind='pie',autopct = '%1.1f%%',colors=['skyblue','lightgreen'])
plt.title('Movies vs. TV Shows on Netflix')
plt.ylabel('')
plt.show()
```

```python
netflix['release_year'].value_counts().sort_index().plot(kind='line',figsize=(8,5),color='red')
plt.title('Netflix Content Added Over the Years')
plt.xlabel('Year')
plt.ylabel('Number of Titles')
plt.grid()
plt.show()
```

```python
# For Movies
movies = netflix[netflix['type'] == 'Movie']
movies['duration'] = movies['duration'].str.replace(' min', '').astype(int)
movies['duration'].plot(kind='hist', bins=20, color='blue', alpha=0.7,rwidth = 0.8)
plt.title('Distribution of Movie Durations (Minutes)')
plt.xlabel('Duration (min)')
plt.show()

# For TV Shows
tv_shows = netflix[netflix['type'] == 'TV Show']
tv_shows['duration'] = tv_shows['duration'].str.replace(' Seasons?', '', regex=True).astype(int)
tv_shows['duration'].value_counts().sort_index().plot(kind='bar', color='green')
plt.title('Number of Seasons in TV Shows')
plt.xlabel('Number of Seasons')
plt.ylabel('Count')
plt.show()
```

```python
from collections import Counter
```

```python
genres = netflix['listed_in'].str.split(', ').explode()
genre_counts = genres.value_counts().head(10)
genre_counts.plot(kind='barh', color='teal')
plt.title('Top 10 Genres on Netflix')
plt.xlabel('Count')
plt.show()
```

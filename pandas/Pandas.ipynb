# Groupby Function

```python
profit = store.groupby('Category')['Profit'].median()
print(profit)
```

# Data Encoding

```python
from sklearn.preprocessing import OneHotEncoder
encoder = OneHotEncoder()
```

```python
encoded = encoder.fit_transform(store[['Category']]).toarray()
```

```python
# numeric dataframe for catergory
df = pd.DataFrame(encoded, columns=encoder.get_feature_names_out())
```
